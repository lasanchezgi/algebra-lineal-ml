# Derivada en ciencia de datos

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2

x = np.linspace(-10,10, num=1000)
y = f(x)

def df(f):
    h = 0.000001
    return (f(x+h)-f(x))/h

plt.plot(x,y, 'b')
plt.plot(x,df(f), 'r')
plt.grid()
plt.show()
```
